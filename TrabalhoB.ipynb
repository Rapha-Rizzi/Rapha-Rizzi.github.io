```python
# Menus.py
import os

class GerenciadorMenus:
    def menuInicial(self):
        os.system('cls')
        print('----------------UNIFLIX - TELA INICIAL----------------')
        print('1 - Acessar')
        print('2 - Criar conta')
        print('0 - Sair da Aplicacao\n')
        item = input('Escolha uma opcao: ')
        return item

    def menuUsuario(self):
        os.system('cls')
        print('----------------MENU DO USUÁRIO----------------')
        print('1 - Alterar a assinatura')
        print('2 - Acessar perfil')
        print('3 - Editar perfil')
        print('4 - Remover perfil')
        print('0 - Voltar para o menu anterior\n')
        item = input('Escolha uma opcao: ')
        return item

    def menuPerfil(self):
        os.system('cls')
        print('----------------MENU DO PERFIL----------------')
        print('1 - Buscar por nome')
        print('2 - Últimos assistidos')
        print('3 - Favoritos')
        print('4 - Filmes')
        print('5 - Séries')
        print('6 - Documentários')
        print('7 - Animações')
        print('8 - Programas de TV')
        print('0 - Voltar para o menu anterior\n')
        item = input('Escolha uma opcao: ')
        return item


# Usuario
class Usuario:
    def __init__(self, id_usuario, nome_usuario, senha, tipo_assinatura):
        self.id_usuario = id_usuario
        self.nome_usuario = nome_usuario
        self.senha = senha
        self.tipo_assinatura = tipo_assinatura
        self.lista_perfis = []

    def adicionarPerfil(self, nome, idade):
         for perfil in self.lista_perfis:
            if perfil.nome == nome:
                print(f"Já existe um perfil com o nome '{nome}'. Escolha outro nome.")
                return

          if len(self.lista_perfis) < self.obterLimitePerfis():

            novo_perfil = Perfil(nome, idade)
            self.lista_perfis.append(novo_perfil)
            print(f"Perfil '{nome}' adicionado com sucesso!")

          else:
            print(f"Limite máximo de perfis atingido para o tipo de assinatura {self.tipo_assinatura}.")

    def removerPerfil(self, nome):
        perfil_encontrado = None
        for perfil in self.lista_perfis:
            if perfil.nome == nome:
                perfil_encontrado = perfil
                break

        if perfil_encontrado:
            self.lista_perfis.remove(perfil_encontrado)
            print(f"Perfil '{nome}' removido com sucesso!")
        else:
            print(f"Perfil '{nome}' não encontrado.")

    def obterLimitePerfis(self):
        return 3 if self.tipo_assinatura == "Simples" else 5


    def alterarSenha(self, nova_senha):
        self.senha = nova_senha

    def alterarPlano(self, novo_tipo):
        self.tipo_assinatura = novo_tipo

    def imprimirInfo(self):
        print(f"ID: {self.id_usuario}")
        print(f"Nome de Usuário: {self.nome_usuario}")
        print(f"Tipo de Assinatura: {self.tipo_assinatura}")
        print("Perfis:")
        for perfil in self.lista_perfis:
            perfil.imprimirInfo()


# Midia
class Midia:
    def __init__(self, tipo, titulo, genero, ano, classificacao):
        self.tipo = tipo
        self.titulo = titulo
        self.genero = genero
        self.ano = ano
        self.classificacao = classificacao

    def exibirInformacoes(self):
        print(f'Tipo: {self.tipo}')
        print(f'Título: {self.titulo}')
        print(f'Gênero: {self.genero}')
        print(f'Ano de Lançamento: {self.ano}')
        print(f'Classificação: {self.classificacao}')


# Serie
class Serie(Midia):
    def __init__(self, titulo, genero, ano, classificacao, num_temporadas, episodios_por_temporada):
        super().__init__('Série', titulo, genero, ano, classificacao)
        self.num_temporadas = num_temporadas
        self.episodios_por_temporada = episodios_por_temporada

    def exibirInformacoes(self):
        super().exibirInformacoes()
        print(f'Número de Temporadas: {self.num_temporadas}')
        print(f'Episódios por Temporada: {self.episodios_por_temporada}')


# Aplicacao
import csv
from Menus import GerenciadorMenus
from Usuario import Usuario
from Serie import Serie

class Aplicacao:
    def __init__(self):
        self.tela = 0
        self.terminou = False
        self.menus = GerenciadorMenus()
        self.listaUsuarios = []
        self.carregarUsuarios('Dados/ExemploUsuarios.csv')
        self.carregarPerfis('ExemploPerfis.csv')
        self.catalogoGeral = []
        self.carregarCatalogo('CatalogoExemplo.csv')
        self.carregarEpisodios('EpisodiosSeriesProgramasTVExemplo.csv')

    def executar(self):
        opcao = -1
        while not self.terminou:
            if self.tela == 0:
                self.telaInicial()
            elif self.tela == 1:
                self.telaMenuUsuario()
            elif self.tela == 2:
                self.telaMenuPerfil()

    def finalizar(self):
        print('Finalizando a aplicacao!')


    def msgErro(self, codigo):
        if codigo == 1:
            print('Opcao invalida!')
            input('Pressione qualquer tecla para continuar')

    def telaInicial(self):
        opcao = self.menus.menuInicial()
        if opcao == '0':
            self.terminou = True
        elif opcao == '1':
            self.fazerLogin()
        elif opcao == '2':
            self.cadastrarUsuario()
        else:
            self.msgErro(1)

    def telaMenuUsuario(self):
        opcao = self.menus.menuUsuario()
        if opcao == '0':
            self.tela = 0
        else:
            self.msgErro(1)

    def telaMenuPerfil(self):
        opcao = self.menus.menuPerfil()
        if opcao == '0':
            self.tela = 1
        else:
            self.msgErro(1)

    def cadastrarUsuario(self):
        print('Cadastrando usuario....')
        input('Pressione qualquer tecla para continuar...')

    def fazerLogin(self):
        print('Verificando usuario e recuperando o album do usuario')
        nome_usuario = input('Digite seu nome de usuário: ')
        senha = input('Digite sua senha: ')
        usuario_encontrado = None
        for usuario in self.listaUsuarios:
            if usuario.nome_usuario == nome_usuario and usuario.senha == senha:
                usuario_encontrado = usuario
                break

        if usuario_encontrado:
            print(f'Login bem-sucedido! Bem-vindo, {usuario_encontrado.nome_usuario}.')
            self.tela = 1
        else:
            print('Nome de usuário ou senha incorretos. Tente novamente.')

        input('Pressione qualquer tecla para continuar...')
        self.tela = 1
        input('Pressione qualquer tecla para continuar...')

    def carregarUsuarios(self, csvUsuarios):
        arqUsuarios = open(csvUsuarios)
        leitor = csv.reader(arqUsuarios, delimiter=';')
        listaCSVUsuarios = list(leitor)
        arqUsuarios.close()

        for i in range(1, len(listaCSVUsuarios)):
            usuario = Usuario(int(listaCSVUsuarios[i][0]), listaCSVUsuarios[i][1], listaCSVUsuarios[i][2],
                              listaCSVUsuarios[i][3])
            self.listaUsuarios.append(usuario)

        self.imprimirInfoUsuarios()

    def imprimirInfoUsuarios(self):
        for usuario in self.listaUsuarios:
            usuario.imprimirInfo()

    def carregarPerfis(self, csvPerfis):
        arqPerfis = open(csvPerfis)
        leitor = csv.reader(arqPerfis, delimiter=';')
        listaCSVPerfis = list(leitor)
        arqPerfis.close()

        for i in range(1, len(listaCSVPerfis)):
            id_usuario = int(listaCSVPerfis[i][0])
            nome_perfil = listaCSVPerfis[i][1]
            idade_perfil = int(listaCSVPerfis[i][2])

            usuario_correspondente = None
            for usuario in self.listaUsuarios:
                if usuario.id_usuario == id_usuario:
                    usuario_correspondente = usuario
                    break


            if usuario_correspondente:

                usuario_correspondente.adicionarPerfil(nome_perfil, idade_perfil)
            else:
                print(f"Erro: Usuário com ID {id_usuario} não encontrado.")




    def carregarCatalogo(self, csvCatalogo):
        arqCatalogo = open(csvCatalogo)
        leitor = csv.reader(arqCatalogo, delimiter=';')
        listaCSVCatalogo = list(leitor)
        arqCatalogo.close()

        for i in range(1, len(listaCSVCatalogo)):
            tipo_midia = listaCSVCatalogo[i][1].lower()
            if tipo_midia == 'serie':
                serie = Serie(titulo=listaCSVCatalogo[i][2], genero=listaCSVCatalogo[i][3],
                              ano=int(listaCSVCatalogo[i][4]), classificacao=listaCSVCatalogo[i][5],
                              num_temporadas=int(listaCSVCatalogo[i][6]), episodios_por_temporada=int(listaCSVCatalogo[i][7]))
                self.catalogoGeral.append(serie)


        self.imprimirInfoCatalogo()

    def imprimirInfoCatalogo(self):
        for midia in self.catalogoGeral:
            midia.exibirInformacoes()

    def carregarEpisodios(self, csvEpisodios):
        arqEpisodios = open(csvEpisodios)
        leitor = csv.reader(arqEpisodios, delimiter=';')
        listaCSVEpisodios = list(leitor)
        arqEpisodios.close()

        for i in range(1, len(listaCSVEpisodios)):
            id_midia = int(listaCSVEpisodios[i][0])
            titulo_temporada = listaCSVEpisodios[i][1]
            numero_temporada = int(listaCSVEpisodios[i][2])
            numero_episodio = int(listaCSVEpisodios[i][3])
            titulo_episodio = listaCSVEpisodios[i][4]


            midia_correspondente = None
            for midia in self.catalogoGeral:
                if midia.id == id_midia:
                    midia_correspondente = midia
                    break

            if midia_correspondente:

                midia_correspondente.adicionarEpisodio(titulo_temporada, numero_temporada, numero_episodio, titulo_episodio)
            else:
                print(f"Erro: Mídia com ID {id_midia} não encontrada.")




if __name__ == "__main__":
    app = Aplicacao()
    app.executar()
    app.finalizar()
```
